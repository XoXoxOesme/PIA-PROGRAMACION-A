```python
class Tres_programas:
    def _init_(self):
        self.compra = 0
        self.confirma = 0
        self.NOM = ""
        self.PES = 0
        self.LON = 0
        self.PESKIL = 0
        self.LONMET = 0
        self.GAL = 0
        self.TOTAL = 0

    def archivos_texto(self):
        with open('descuento.txt', 'w') as file:
            if self.compra < 500:
                file.write("No hay descuento\n")
            else:
                if 500 <= self.compra < 1000:
                    descuento = 0.05 * self.compra
                elif 1000 <= self.compra < 7000:
                    descuento = 0.11 * self.compra
                elif 7000 <= self.compra < 15000:
                    descuento = 0.15 * self.compra
                else:
                    descuento = 0.20 * self.compra

                self.compra -= descuento
                file.write(f"El total es {self.compra}\n")

        with open('pago.txt', 'w') as file:
            pago = self.confirma - self.compra
            file.write(f"Su descuento es {pago}\n")

        with open('dinosaurio.txt', 'w') as file:
            file.write(f"Nombre del dinosaurio: {self.NOM}\n")
            file.write(f"Peso en kilogramos: {self.PESKIL} kg\n")
            file.write(f"Longitud en metros: {self.LONMET} m\n")

        with open('gasolinera.txt', 'w') as file:
            litros = self.GAL * 3.785
            self.TOTAL = litros * 8.20
            file.write(f"Galones surtidos: {self.GAL} galones\n")
            file.write(f"Total a pagar: ${self.TOTAL:.2f}\n")

    def descuento(self):
        if self.compra < 500:
            return "No hay descuento"
        else:
            if 500 <= self.compra < 1000:
                descuento = 0.05 * self.compra
            elif 1000 <= self.compra < 7000:
                descuento = 0.11 * self.compra
            elif 7000 <= self.compra < 15000:
                descuento = 0.15 * self.compra
            else:
                descuento = 0.20 * self.compra

            self.compra -= descuento
            return f"El total es {self.compra}"

    def pago(self):
        pago = self.confirma - self.compra
        return f"Su descuento es {pago}"

    def dinosaurio(self, NOM, PES, LON):
        self.NOM = NOM
        self.PES = PES
        self.LON = LON
        self.PESKIL = PES / 2.20462
        self.LONMET = LON * 0.3047

    def imprimir_datos(self):
        return f"Nombre del dinosaurio: {self.NOM}\nPeso en kilogramos: {self.PESKIL} kg\nLongitud en metros: {self.LONMET} m"

    def gasolinera(self):
        self.GAL = float(input("Ingrese la cantidad de galones surtidos: "))

    def gasolina(self):
        litros = self.GAL * 3.785
        self.TOTAL = litros * 8.20

    def imprimir_resultados(self):
        return f"Galones surtidos: {self.GAL} galones\nTotal a pagar: ${self.TOTAL:.2f}"


# Bloque principal
Tres_programas1 = Tres_programas()
Tres_programas1.compra = float(input("Ingrese la cantidad a pagar: "))
Tres_programas1.confirma = float(input("Confirme la cantidad a pagar: "))

resultado_descuento = Tres_programas1.descuento()
resultado_pago = Tres_programas1.pago()

# Segundo programa
NOM_dino = input("Ingrese el nombre del dinosaurio: ")
PES_dino = float(input("Ingrese el peso del dinosaurio en libras: "))
LON_dino = float(input("Ingrese la longitud del dinosaurio en pies: "))

Tres_programas1.dinosaurio(NOM_dino, PES_dino, LON_dino)
resultado_dinosaurio = Tres_programas1.imprimir_datos()

# Tercer programa
Tres_programas1.gasolinera()
Tres_programas1.gasolina()
resultado_gasolinera = Tres_programas1.imprimir_resultados()

# Imprimir resultados
print(resultado_descuento)
print(resultado_pago)
print(resultado_dinosaurio)
print(resultado_gasolinera)

# Generar archivos
Tres_programas1.archivos_texto()
```
